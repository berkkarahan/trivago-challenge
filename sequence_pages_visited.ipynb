```python
def simple_hashmap(obj, mapsize=100):
    h = hash(obj)
    return h % mapsize
```

```python
indices = df.index.values
```

```python
hsize = 100
hmap = np.zeros((sh[0],hsize))
```

```python
idx = 0
for row in df.itertuples():
    path = getattr(row, 'path_id_set')
    path = path.split(';')
    for s in path:
        i = simple_hashmap(s, hsize)
        hmap[idx, i] = hmap[idx, i] + 1
    idx = idx + 1
```
